```python
import os
from glob import glob
from tqdm import tqdm
from util import get_parent_dir_name, read_json_as_dict
import shutil
```

```python
train_json_path = "../data/annotations/person_keypoints_train2017.json"
valid_json_path = "../data/annotations/person_keypoints_val2017.json"
single_image_folder_path = "../data/1. Single Person"
os.makedirs(f"{single_image_folder_path}/train", exist_ok=True)
os.makedirs(f"{single_image_folder_path}/val", exist_ok=True)
train_image_folder_path = "../data/train2017"
valid_image_folder_path = "../data/val2017"
```

```python
train_annotation_dict = read_json_as_dict(train_json_path)
valid_annotation_dict = read_json_as_dict(valid_json_path)
```

# Collect Single Person Data

```python
def get_filtered_dict(total_annotation_list):
    image_id_dict = {}
    for annotation_dict in total_annotation_list['annotations']:
        image_id = annotation_dict['image_id']
        num_keypoints = annotation_dict['num_keypoints']
        if num_keypoints > 0:
            if image_id in image_id_dict:
                image_id_dict[image_id].append(annotation_dict)
            else:
                image_id_dict[image_id] = [annotation_dict]

    single_image_id_dict = {
        key: value for key, value in image_id_dict.items() if len(value) == 1
    }
    multi_image_id_dict = {
        key: value for key, value in image_id_dict.items() if len(value) > 1
    }    
        
    return single_image_id_dict, multi_image_id_dict
```

```python
train_single_image_id_dict, train_multi_image_id_dict = get_filtered_dict(train_annotation_dict)
valid_single_image_id_dict, valid_multi_image_id_dict = get_filtered_dict(valid_annotation_dict)
```

```python
train_image_path_list = glob(f"{train_image_folder_path}/*.*")
valid_image_path_list = glob(f"{valid_image_folder_path}/*.*")
```

```python
for image_path in tqdm(train_image_path_list):
    image_basename = get_parent_dir_name(image_path, level=0) 
    image_id = int(image_basename.split(".")[0])
    
    if image_id in train_single_image_id_dict.keys():
        phase = "single_person"
    elif image_id in train_multi_image_id_dict.keys():
        phase = "multi_persion"
    else:
        phase = "no_person"
    
    target_image_path = f"{single_image_folder_path}/train/{phase}/{image_basename}"
    shutil.copy(image_path, target_image_path)
```

```python
for image_path in tqdm(valid_image_path_list):
    image_basename = get_parent_dir_name(image_path, level=0) 
    image_id = int(image_basename.split(".")[0])
    
    if image_id in train_single_image_id_dict.keys():
        phase = "single_person"
    elif image_id in train_multi_image_id_dict.keys():
        phase = "multi_persion"
    else:
        phase = "no_person"
    
    target_image_path = f"{single_image_folder_path}/train/{phase}/{image_basename}"
    shutil.copy(image_path, target_image_path)
```

```python
import json

with open(f'{single_image_folder_path}/single_person_keypoints_train2017.json', 'w') as fp:
    json.dump(train_single_image_id_dict, fp)
with open(f'{single_image_folder_path}/multi_person_keypoints_train2017.json', 'w') as fp:
    json.dump(train_multi_image_id_dict, fp)
with open(f'{single_image_folder_path}/single_person_keypoints_val2017.json', 'w') as fp:
    json.dump(valid_single_image_id_dict, fp)
with open(f'{single_image_folder_path}/multi_person_keypoints_val2017.json', 'w') as fp:
    json.dump(valid_multi_image_id_dict, fp)
```
